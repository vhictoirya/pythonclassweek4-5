# Understanding Blockchain Data Structures

## Blocks & Transactions: The Ledger of Blockchain

### Blocks are the fundamental units of a blockchain:
- Contain batches of validated transactions

- Linked cryptographically in chronological order

- Include metadata like timestamp, miner, and difficulty

### Transactions represent actions on the blockchain:
- Value transfers (sending ETH/tokens)

- Smart contract interactions

### Contain critical fields:
- from/to addresses
- value transferred
- input data (for contracts)
- gas parameters

```python
# Using web3.py too get block details

from web3 import Web3  

w3 = Web3(Web3.HTTPProvider('YOUR_PROVIDER_URL'))  

block = w3.eth.get_block('latest')  

print(f"Block #{block['number']} has {len(block['transactions'])} transactions")  
```

## Gas & Fee Structure: Blockchain's Resource Economics

### Gas measures computational work:
- Each operation has a fixed gas cost
- Simple transfer: 21,000 gas (might vary)
- Complex contract calls: 100,000+ gas (might vary)

### Fee Calculation

```python
Total Fee = gas_used * gas_price

# gas_price: Price per unit (set by user, in Gwei)
# gas_used: Actual units consumed
```

## State & Storage: Smart Contract Memory

### State represents current data:
- Account balances
- Smart contract variables
- Stored in Merkle Patricia Tries

### Storage specifics:
- Each contract has 2²⁵⁶ storage slots
- Complex data structures use hashed positions
- State changes only occur in transactions
